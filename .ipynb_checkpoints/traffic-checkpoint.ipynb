```python
import pandas as pd
import matplotlib.pyplot as plt

#load the dataset
traffic = pd.read_csv("Traffic_Data.csv")
print(traffic)
```

```python
print(traffic.info())
#Find missing if there missing values
print(traffic.isnull().sum())
```

```python
traffic['Day Of Week'] = pd.Categorical(traffic['Day Of Week'], categories=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], ordered=True)
traffic_density_by_day = traffic.groupby('Day Of Week')['Traffic Density'].mean()

# Plotting traffic density by day of the week
plt.figure(figsize=(10, 6))
traffic_density_by_day.plot(kind='bar', color='skyblue')
plt.title('Average Traffic Density by Day of the Week')
plt.xlabel('Day Of Week')
plt.ylabel('Average Traffic Density')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
