#### Insights on landslide occurrences in Iran

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
landslide_data = pd.read_csv('Landslide_Factors_IRAN.csv')

# Display basic information about the dataset
print(landslide_data.info())

# Check for missing values
print(landslide_data.isnull().sum())

# Remove duplicates
landslide_data.drop_duplicates(inplace=True)
```

```python
# Visualize the distribution of landslide occurrences by land use type
plt.figure(figsize=(10, 6))
sns.countplot(x='Landuse_Type', data=landslide_data)
plt.title('Distribution of Landslide Occurrences by Land Use Type')
plt.xlabel('Land Use Type')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right', fontsize=10)  
plt.tight_layout()  
plt.show()

# The relationship between slope and landslide occurrence by land use type
plt.figure(figsize=(10, 6))
sns.boxplot(x='Landuse_Type', y='Slop(Percent)', data=landslide_data)
plt.title('Land Use Type vs. Slope (Percentage)')
plt.xlabel('Land Use Type')
plt.ylabel('Slope (Percentage)')
plt.xticks(rotation=45, ha='right', fontsize=10) 
plt.tight_layout()  
plt.show()

# The distribution of landslide occurrences by climate type
plt.figure(figsize=(10, 6))
sns.countplot(x='Climate_Type', data=landslide_data)
plt.title('Distribution of Landslide Occurrences by Climate Type')
plt.xlabel('Climate Type')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right', fontsize=10) 
plt.tight_layout()  
plt.show()
```

#### A dashboard to exhibit interpretation of the data analytics.

```python

# Dashboard-like layout with increased figure size
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Plot 1: Distribution of AAP(mm)
sns.histplot(data=landslide_data, x='AAP(mm)', bins=20, kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Distribution of AAP (mm)')
axes[0, 0].set_xlabel('AAP (mm)')
axes[0, 0].set_ylabel('Frequency')

# Plot 2: AAP vs. Elevation by Landuse Type
sns.scatterplot(data=landslide_data, x='Elevation', y='AAP(mm)', hue='Landuse_Type', ax=axes[0, 1])
axes[0, 1].set_title('AAP vs. Elevation by Landuse Type')
axes[0, 1].set_xlabel('Elevation')
axes[0, 1].set_ylabel('AAP (mm)')
axes[0, 1].legend(title='Landuse Type')

# Plot 3: AAP vs. Slop(Percent) by GEO_UNIT
sns.scatterplot(data=landslide_data, x='Slop(Percent)', y='AAP(mm)', hue='GEO_UNIT', ax=axes[1, 0])
axes[1, 0].set_title('AAP vs. Slop(Percent) by GEO_UNIT')
axes[1, 0].set_xlabel('Slop(Percent)')
axes[1, 0].set_ylabel('AAP (mm)')
axes[1, 0].legend(title='GEO_UNIT')

# You can add more plots here...

plt.tight_layout()
plt.show()
```
